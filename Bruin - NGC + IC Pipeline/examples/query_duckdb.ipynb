# DuckDB example queries

This notebook demonstrates simple queries against the pipeline DuckDB warehouse. Update the `DB_PATH` below if your warehouse file is in a different location.

```python
import duckdb
import pandas as pd

# Path to the warehouse DuckDB file (relative to this notebook)
DB_PATH = 'warehouse/bruin.duckdb'

con = duckdb.connect(DB_PATH)

# Example 1: count objects by normalized type in NGC catalog
q1 = '''
SELECT normalized_type, COUNT(*) AS cnt
FROM ngc_catalog
GROUP BY 1
ORDER BY cnt DESC
LIMIT 100
'''
df1 = con.execute(q1).df()
print(df1)

# Example 2: sample rows with coordinates and types from both catalogs
q2 = '''
SELECT ngc.NGC, ngc.Name AS ngc_name, ngc.RA AS ngc_ra, ngc.DEC AS ngc_dec, ngc.normalized_type AS ngc_type,
       ic.IC, ic.Name AS ic_name, ic.RA AS ic_ra, ic.DEC AS ic_dec, ic.normalized_type AS ic_type
FROM ngc_catalog ngc
LEFT JOIN ic_catalog ic ON ngc.constellation = ic.constellation
LIMIT 50
'''
df2 = con.execute(q2).df()
df2.head()

# Close connection
con.close()
```
